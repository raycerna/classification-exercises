```python
def prep_iris(iris_df):
    '''
    Takes in a iris dataframe and returns a cleaned dataframe
    Arguments: df - a pandas dataframe with the expected feature names and columns
    Return: clean_df - a dataframe with the cleaning operations performed on it
    '''

    #Drop Columns
    columns_to_drop = ['species_id', 'measurement_id']
    iris_df = iris_df.drop(columns = columns_to_drop)
    #renamed column
    iris_df = iris_df.rename(columns={'species_name': 'species'})
    #Encoded categorical variables
    dummy_df = pd.get_dummies(iris_df[['species']], dummy_na=False)
    iris_df = pd.concat([iris_df, dummy_df], axis=1)
    return iris_df
```

```python
iris_df = acquire.get_iris_data()
iris_df = prep_iris(iris_df)
iris_df.head()
```

```python
def prep_titanic(titanic_df):
    '''
    Takes in a titanic datafram and returns a cleaned dataframe
    Arguments: df - a pandas dataframe with the expected feature names and columns
    Return: clean_df - a dataframe with the cleaning operations performed on it
    '''
    #Drop duplicates
    titanic_df.drop_duplicates(inplace=True)
    #Drop Columns
    columns_to_drop = ['embarked', 'class', 'passenger_id', 'deck']
    titanic_df = titanic_df.drop(columns = columns_to_drop)
    #Encoded categorical variables
    dummy_df = pd.get_dummies(titanic_df[['sex', 'embark_town']], dummy_na=False, drop_first=[True, True])
    titanic_df = pd.concat([titanic_df, dummy_df], axis=1)
    return titanic_df.drop(columns=['sex', 'embark_town'])
```

```python
titanic_df = acquire.get_titanic_data()
titanic_df = prep_titanic(titanic_df)
titanic_df.head()
```
